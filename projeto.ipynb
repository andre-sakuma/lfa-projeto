```python
from automathon import DFA

Q = {'abrindo', 'aberto', 'travado', 'fechando', 'fechado'}
sigma = {'0', '1'}
delta = { 'abrindo' : {'0' : 'abrindo', '1' : 'aberto'}, # porta totalmente aberta?
          'aberto' : {'0' : 'aberto', '1' : 'fechando'}, # passaram-se 60 segundos?
          'fechando': {'0': 'travado', '1': 'fechado'}, # porta totalmente fechada?
          'travado' : {'0' : 'travado', '1' : 'fechando'}, # passaram-se 10 segundos?
          'fechado': {'0': 'fechado', '1': 'abrindo'} # parou na esta√ß√£o?
        }
initial_state = 'fechado'
F = {}

automata = DFA(Q, sigma, delta, initial_state, F)
```

```python
automata.is_valid()
```

 **`Visualiza√ß√£o`**

```python
# estilo default
automata.view("afd01")
```

```python
# estilo personalizado
automata.view(
    file_name="afd01_personalizado",
    node_attr={'fontsize': '20'},
    edge_attr={'fontsize': '20pt'}
)
```
